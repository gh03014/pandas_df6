```python
#fas-정재환
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import numpy as np
import csv
import json
import seaborn as sns
%matplotlib inline 
```

```python
# 새로 생성할 파일 이름 리스트
f_list = ['건강검진1', 
          '건강검진2', 
          '건강검진3']
```

```python
# 원본 데이터의 1~50000번째 행의 5만개 데이터를 건강검진1파일에 저장
# 원본 데이터의 50000~100000번째 행의 5만개 데이터를 건강검진2파일에 저장
# 원본 데이터의 100000~150000번째 행의 5만개 데이터를 건강검진3파일에 저장
#skiprows를 이용해 지정된 행은 생략한 후에 시작 오픈셋을 만든다.
#nrows를 이용해 위에서부터 지정한 숫자 만큼의 행만 읽어온다
for i in range(len(f_list)):
    df_columns  = pd.read_csv('NHIS_OPEN_GJ_2017_2.csv', encoding='cp949', nrows=0)
    df_values = pd.read_csv('NHIS_OPEN_GJ_2017_2.csv', encoding='cp949', skiprows=(i*50000), nrows=50000, 
                           names = df_columns.columns, header = 0)
    df_values.reset_index(drop=True, inplace=True)
    df_values.to_csv(f_list[i] + '.csv', mode='w', encoding='cp949')
    del df_columns # 메로리 반환, 누수 방지
    del df_values
```

```python
# 3개의 생성된 csv 파일 확인
```

```python
#건강검진1.csv 확인
df1 = pd.read_csv(f_list[0] + '.csv', encoding='cp949', index_col=0)
df1
```

```python
#건강검진2.csv 확인
df2 = pd.read_csv(f_list[1] + '.csv', encoding='cp949', index_col=0)
df2
```

```python
#건강검진3.csv 확인
df3 = pd.read_csv(f_list[2] + '.csv', encoding='cp949', index_col=0)
df3
```

```python
# concat을 이용해 건강검진1, 2, 3의 3개 파일 병합
df = pd.concat([df1, df2], axis=0)
df = pd.concat([df, df3], axis=0)
del df1 # 메로리 반환, 누수 방지
del df2
del df3
df
```

```python
del df # 메로리 반환, 누수 방지
```

```python
# 이번에는 다른 방법으로 3개의 파일을 하나의 파일에 이어쓰기로 저장, 병합하고
# 병합된 csv 파일을 읽어서 데이터프레임으로 변환한다
```

```python
# 먼저 컬럼명만 가져와서 테이블의 구조만 생성한다
df_columns  = pd.read_csv('건강검진1.csv', encoding='cp949', nrows=0, index_col=0)
df_columns.to_csv('merge_건강검진.csv', mode='w', encoding='cp949')
del df_columns
# 반복문을 통해 모든 파일을 읽어서 csv파일에 이어쓰기(모드: a)한다.
#메모리 누적 오류를 방지하기 위해 한번 읽을때 마다 del을 이용해 메모리를 반환한다.
for i in range(len(f_list)):
    df = pd.read_csv(f_list[i] + '.csv', encoding='cp949')
    df.to_csv('merge_건강검진.csv', mode='a', header=False, index=True)
    del df
```
